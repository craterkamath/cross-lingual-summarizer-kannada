## Comparison between rouge1 rouge2 and rougel

```python
df_summary = pd.DataFrame(columns=["rouge1","rouge2","rougel"])
```

```python
df_summary["rouge1"]=df_rouge1["summary"]
df_summary["rouge2"]=df_rouge2["summary"]
df_summary["rougel"]=df_rougel["summary"]
```

```python
df_summary.head()
```

```python
sns.pairplot(df_summary)
```

```python
sns.kdeplot(df_summary["rouge1"], label = "rouge1")
sns.kdeplot(df_summary["rouge2"], label = "rouge2")
sns.kdeplot(df_summary["rougel"], label = "rougel")
plt.legend()
plt.title("Summary Rogue Score Distributions")
plt.show()
```
